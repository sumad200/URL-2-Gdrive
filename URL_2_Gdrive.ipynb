```python
import os
from google.colab import files
uploaded = files.upload()
if(len(uploaded)!=1):
  raise Exception("Please re-run this cell and Upload a single text file with each link in a new line")
first_key = list(uploaded.keys())[0]
split_tup = os.path.splitext(first_key)
fileext = split_tup[1]
if(fileext != '.txt'):
    raise Exception("Please re-run this cell and Upload a single text file with each link in a new line")
```

```python
import os
inputfile = open(first_key,'r',encoding='utf-8')
data = inputfile.read().replace('\r\n', '\n')
open('links.txt','w',encoding='utf-8').write(data)
```

```python
%%shell

urllink='/content/links.txt'
downpath='/content/drive/MyDrive/Downloads/'
Now_hourly=$(date +%d-%m-%Y)  
echo "$Now_hourly"
downfolder='download-'
downfolder+="$Now_hourly"
downpath+="$downfolder"
mkdir -p $downpath
while IFS= read -r line
do
  
  echo "$line"
  wget $line -P $downpath
done < "$urllink"
```
